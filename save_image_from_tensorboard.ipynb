```python
import tensorflow as tf

def list_image_tags(event_file):
    tags = set()
    for e in tf.compat.v1.train.summary_iterator(event_file):
        for v in e.summary.value:
            if v.image.ByteSize():  # Check if it's an image type summary
                tags.add(v.tag)
    return tags

event_file = '/home/yuechuan/vlr/project/neu-nbv/scripts/neural_rendering/logs/urbanscene_training_cutscene_60/events.out.tfevents.1702524799.yuechuan-Ubuntu20.741685.0'
print(list_image_tags(event_file))
```

```python
import os
from PIL import Image
import numpy as np
import tensorflow as tf

def save_images_from_event(fn, tag, output_dir='./'):
    # Create the output directory if it does not exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Iterate over events in the file
    for e in tf.compat.v1.train.summary_iterator(fn):
        for v in e.summary.value:
            if v.tag == tag:
                # Decode the image
                im = tf.io.decode_image(v.image.encoded_image_string)
                # Convert to numpy array
                im_array = np.array(im)
                # Generate output file name
                output_fn = os.path.join(output_dir, f'image_{e.step}.png')
                # Save the image using PIL
                Image.fromarray(im_array).save(output_fn)
                print(f"Saving '{output_fn}'")

# Example usage
event_file_path = '/home/yuechuan/vlr/project/neu-nbv/scripts/neural_rendering/logs/urbanscene_training_cutscene_60/events.out.tfevents.1702524799.yuechuan-Ubuntu20.741685.0'
chosen_tag = 'predict/uncertainty'
output_directory = './urbanscene_training_cutscene_60/predict_uncertainty'  

save_images_from_event(event_file_path, chosen_tag, output_directory)
```

```python
import tensorflow as tf

def list_scalar_tags(event_file):
    scalar_tags = set()
    for e in tf.compat.v1.train.summary_iterator(event_file):
        for v in e.summary.value:
            if v.metadata.plugin_data.plugin_name == 'scalars':
                scalar_tags.add(v.tag)
    return scalar_tags

# Replace with your actual event file path
event_file_path = '/home/yuechuan/vlr/project/neu-nbv/scripts/neural_rendering/logs/urbanscene_training_90/events.out.tfevents.1701772022.yuechuan-Ubuntu20.3729525.0'
tags = list_scalar_tags(event_file_path)
print(tags)
```
